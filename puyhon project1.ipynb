```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

import pandas as pd
import numpy as np
data = pd.read_csv('dataforprojectog.csv')
data
```

```python
data['Height']= np.random.uniform(150, 180, size = len(data))
data
data.drop_duplicates()
```

```python
v = data['Salary'].mean()
v
```

```python
data['Salary'].fillna(v, inplace= True)
data
```

```python
w = data['Weight'].mean()
data['Weight'].fillna(w, inplace= True)
data
```

```python
A = data['Age'].mean()
data['Age'].fillna(A, inplace= True)
data
```

```python
N = data['Number'].mean()
data['Number'].fillna(N, inplace= True)
data
```

```python
data.isnull().sum()
```

```python
data.to_csv('cleaneddata.csv', index=False)
```

```python
Find from which age group most of the employees belong to
```

```python
age_group= pd.cut(data['Age'], bins=[20, 25, 30, 35, 40], labels=['20-24', '25-29', '30-34', '35-39'])

age_group
```

```python
count_age= age_group.value_counts()
count_age
```

```python
print("age group of most employers=", count_age.idxmax())
```

```python
#Find out under which team and position, spending in terms of salary is high.
```

```python

g_data = data.groupby(['Team', 'Position'])


salary_t_p = g_data['Salary'].sum()
salary_t_p.idxmax()


print("highest paid team by position", salary_t_p.idxmax())
```

```python
#Find if there is any correlation between age and salary , represent it visually.
```

```python

correlation = data['Age'].corr(data['Salary'])
print('Correlation coefficient between age and salary:', correlation)

plt.scatter(data['Age'], data['Salary'])
plt.xlabel('Age (years)')
plt.ylabel('Salary (USD)')
plt.title('Age vs Salary')
plt.show()
```
